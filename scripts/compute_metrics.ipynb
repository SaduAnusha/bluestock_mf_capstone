```python
"""
BLUESTOCK MF - Compute Metrics
"""
import numpy as np
from scipy import stats

def compute_sharpe(returns):
    """Calculate annualized Sharpe ratio"""
    return (returns.mean() * 252 - 0.065) / (returns.std() * np.sqrt(252))

def compute_alpha_beta(fund, bench):
    """Compute Alpha and Beta"""
    slope, intercept, _, _, _ = stats.linregress(bench, fund)
    return intercept * 252, slope

def compute_var(returns):
    """Calculate VaR at 95%"""
    return np.percentile(returns, 5)

if __name__ == "__main__":
    print("Compute Metrics Module Ready!")
```
