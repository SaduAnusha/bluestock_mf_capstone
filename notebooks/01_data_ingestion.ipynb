```python
import pandas as pd
from pathlib import Path

raw_path = Path("../data/raw/Bluestock_MF_Datasets")

files = list(raw_path.glob("*.csv"))

print(files)
print("Total files:", len(files))
```

```python
datasets = {}

for file in files:
    df = pd.read_csv(file)
    datasets[file.name] = df

    print("\n" + "="*50)
    print("File:", file.name)
    print("Shape:", df.shape)
    print("\nData Types:")
    print(df.dtypes)
    print("\nFirst 5 Rows:")
    print(df.head())
```

```python
fund_master = datasets["01_fund_master.csv"]

print("\n" + "="*50)
print("FUND MASTER EXPLORATION")
print("="*50)

print(f"\nUnique Fund Houses: {fund_master['fund_house'].nunique()}")
print(fund_master['fund_house'].unique())

print(f"\nCategories: {fund_master['category'].unique()}")

print(f"\nSub Categories: {fund_master['sub_category'].unique()}")

print(f"\nRisk Categories: {fund_master['risk_category'].unique()}")
```

```python
nav_codes = set(datasets["02_nav_history.csv"]["amfi_code"].unique())
master_codes = set(datasets["01_fund_master.csv"]["amfi_code"].unique())

missing_codes = nav_codes - master_codes

print("NAV Codes:", len(nav_codes))
print("Fund Master Codes:", len(master_codes))

if len(missing_codes) == 0:
    print("✅ All AMFI codes are valid!")
else:
    print("❌ Missing Codes:", missing_codes)
```
